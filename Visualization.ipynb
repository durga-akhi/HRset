```python

plt.figure(figsize=(8, 6))
sns.countplot(data=dt, x='Business Travel', hue='Attrition')
plt.title('Attrition by Business Travel')
plt.xlabel('Business Travel')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Attrition')
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.boxplot(data=dt, x='CF_age band', y='Age', hue='Attrition')
plt.title('Age Distribution by Attrition and Age Band')
plt.xlabel('Age Band')
plt.ylabel('Age')
plt.xticks(rotation=45)
plt.legend(title='Attrition')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=dt, x='Department', y='Monthly Income', hue='Attrition', estimator=sum)
plt.title('Total Monthly Income by Department and Attrition')
plt.xlabel('Department')
plt.ylabel('Total Monthly Income')
plt.xticks(rotation=45)
plt.legend(title='Attrition')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=dt, x='Attrition', y='Total Working Years')
plt.title('Total Working Years by Attrition')
plt.xlabel('Attrition')
plt.ylabel('Total Working Years')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=dt, x='Attrition', y='Years At Company')
plt.title('Years At Company by Attrition')
plt.xlabel('Attrition')
plt.ylabel('Years At Company')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=dt, x='Total Working Years', y='Monthly Income', hue='Attrition')
plt.title('Monthly Income vs Total Working Years by Attrition')
plt.xlabel('Total Working Years')
plt.ylabel('Monthly Income')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=dt, x='Age', hue='Attrition', multiple='stack', kde=True)
plt.title('Age Distribution by Attrition')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Attrition')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=dt, x='Department', y='Performance Rating')
plt.title('Performance Rating by Department')
plt.xlabel('Department')
plt.ylabel('Performance Rating')
plt.tight_layout()
plt.show()
```
